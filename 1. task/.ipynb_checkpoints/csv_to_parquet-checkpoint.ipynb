```python
from pyspark.sql import SparkSession
```

```python
#1. create sparksession
spark = SparkSession.builder.appName("CSV to Parquet data").getOrCreate()
```

```python
# 2. Load CSV file

df = spark.read.csv(
    r"D:\Kadel Labs Training\1. task\customers.csv",
    header=True,
    inferSchema=True
)

df.show()
```

```python
# 3. Display Schema
print("Schema of DataFrame:")
df.printSchema()
```

```python
# 4. Preview Data (Top 20 rows default)
print("Preview Data:")
df.show()
```

```python
# 5. Save DataFrame as Parquet
output_path = "output_parquet"

df.write.mode("overwrite").parquet(output_path)

print("Data saved successfully in Parquet format.")
```
